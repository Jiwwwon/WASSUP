# Label Encoding 
문자형 변수를 숫자형 변수로 인코딩

```python
import warnings
warnings.filterwarnings('ignore')
```

# 실습 데이터  
자동차 연비(mpg) 데이터
- mpg : miles per gallon
- 1 mile : 1.6 Km
- 1 gallon : 3.78 Liter

```python
import seaborn as sns

DF = sns.load_dataset('mpg')
```

```python
DF.info()
```

```python
DF.head()
```

```python
type(DF.origin[0])  # 문자형 데이터 : 'origin'
```

```python
DF.origin.value_counts()
```

```python
X = DF[['origin']]
```

```python
X[111:115]
```

# With LabelEncoder 
정수(Integer) 인코딩

```python
from sklearn.preprocessing import LabelEncoder

encoder1 = LabelEncoder()
LE = encoder1.fit_transform(X)
```

```python
LE[111:115]
```

# With OneHotEncoder 
원-핫(One-Hot) 인코딩

```python
from sklearn.preprocessing import OneHotEncoder 

encoder2 = OneHotEncoder()
OHE = encoder2.fit_transform(X)
```

```python
print(OHE[111:115])
```

```python
OHE.toarray()[111:115]
```

# Scaling 
범위(Scale)가 다른 변수의 범위(Scale)를 비슷하게 맞추기 위한 목적

```python
import warnings
warnings.filterwarnings('ignore')
```

# 실습 데이터

```python
import seaborn as sns
DF = sns.load_dataset('mpg')
```

자동차 연비(mpg) 데이터
- mpg : miles per gallon
- 1 mile : 1.6 Km
- 1 gallon : 3.78 Liter

```python
DF.info()
```

```python
DF.head()
```

```python
X = DF[['weight']]
y = DF['mpg']
```

# Without Scaling

```python
X[:5]
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (9, 6))
sns.regplot(x = X, y = y)
plt.xlabel('weight_Without_Scaling')
plt.show() # 회귀선 동일하게 도출
```

# With Normalization

```python
from sklearn.preprocessing import MinMaxScaler

scaler1 = MinMaxScaler()
X_Norm = scaler1.fit_transform(X)
```

## 정규화된 X값 확인

```python
X_Norm[:5]
```

```python
fig = plt.figure(figsize = (9, 6))
sns.regplot(x = X_Norm, y = y)
plt.xlabel('weight_With_Normalization')
plt.show()
```

# With Standardization

```python
from sklearn.preprocessing import StandardScaler

scaler2 = StandardScaler()
X_Stan = scaler2.fit_transform(X)
```

## 표준화된 X값 확인

```python
X_Stan[:5]
```

```python
fig = plt.figure(figsize = (9, 6))
sns.regplot(x = X_Stan, y = y)
plt.xlabel('weigh_With_Standardization')
plt.show()
```
